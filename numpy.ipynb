```python
arr=np.arange(100)
filter=arr[(arr>20)&(arr<40)]
filter
```

```python
import pandas as pd
```

```python
dataset=pd.DataFrame([arr,filter])
dataset
```

```python
# removing NaN values from the data set assigning 0 in those position
clean_data=pd.DataFrame(dataset.fillna(0))
clean_data
```

```python
import matplotlib.pyplot as plt

```

```python
plt.plot(dataset,clean_data[::-1],"*")
plt.show()
```
