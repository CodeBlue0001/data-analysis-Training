```python
# loading dataset from csv
Dataset=pd.read_csv("/content/pets (1).csv - pets (1).csv")
Dataset
```

```python
# separeting dogs dataset
dog_datset=pd.DataFrame(Dataset[Dataset["Pets"]=="dog"])
dog_datset
```

```python
# cleaning Nan data from the dog dataset
clean_dog_dataset=pd.DataFrame(dog_datset.fillna(0))
clean_dog_dataset
```

```python
# taking priority yes dogs
priority_dog=clean_dog_dataset[clean_dog_dataset["Priority"]=="Yes"]
priority_dog
```

```python
# dogs and cats in the plot

cat_dataset=pd.DataFrame(Dataset[Dataset["Pets"]=="cat"])
cat_dataset
min_len = min(len(clean_dog_dataset["Visits"]), len(cat_dataset["Visits"]))
plt.xlabel("Visits")
plt.ylabel("Age")
# ploting dogs data
plt.plot(clean_dog_dataset["Visits"].iloc[:min_len],clean_dog_dataset["Age"].iloc[:min_len])
# ploting cat data on this
plt.plot(cat_dataset["Visits"].iloc[:min_len],cat_dataset["Age"].iloc[:min_len])
plt.show()
```
