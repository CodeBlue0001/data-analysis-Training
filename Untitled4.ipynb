```python
import pandas as pd

# Convert 'sales' column to numeric, coercing errors to NaN
df['sales'] = pd.to_numeric(df['sales'], errors='coerce')

# Create a DataFrame with 'sales' and 'country' as columns
country_sales = df[['sales', 'country']]

# Display the first few rows to verify the structure
print("First 5 records of country_sales:")
print(country_sales.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

country=country_sales["country"]
sales=country_sales['sales']

plt.bar(country,sales)
# plt.bar_label
x=plt.xticks(rotation=90)
# make a gap between bars to display the names properly
# plt.subplots_adjust(left=10)

plt.xlabel('country')
plt.ylabel('sales')
plt.show()
```

```python
# using seaborn

sns.barplot(x=country,y=sales,data=country_sales)
```
