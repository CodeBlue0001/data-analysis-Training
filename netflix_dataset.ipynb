```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv('/content/netflix_titles.csv')


# We are filling in the missing text data with 'Unknown' instead of deleting the rows
df['director'] = df['director'].fillna('Unknown')
df['country'] = df['country'].fillna('Unknown')
df['cast'] = df['cast'].fillna('Unknown')


plt.figure(figsize=(8, 5))
type_counts = df['type'].value_counts()

```

```python
# Replacing missing text with 'Unknown'
df['director'] = df['director'].fillna('Unknown')
df['country'] = df['country'].fillna('Unknown')
df['cast'] = df['cast'].fillna('Unknown')
df['date_added'] = df['date_added'].fillna('Unknown')

# For ratings and durations, 'Not Rated' or 'Unknown'
df['rating'] = df['rating'].fillna('Not Rated')
df['duration'] = df['duration'].fillna('Unknown')

print("\nChecking for Missing Values (After Cleaning):")
print(df.isnull().sum())
print("Data is 100% clean and ready for visualization!\n")
print("-" * 50)







```

```python
# CHART 1: The "Movie vs TV Show" Split (Pie Chart)
plt.figure(figsize=(8, 5))
type_counts = df['type'].value_counts()
plt.pie(type_counts, labels=type_counts.index, autopct='%1.1f%%',
        colors=['#E50914', '#564d4d'], startangle=90, textprops={'fontsize': 12})
plt.title('Netflix Catalog: Movies vs. TV Shows', fontsize=16, fontweight='bold')
plt.show()
```

```python

plt.figure(figsize=(10, 6))
# We grab only the top 10 ratings so the chart doesn't look cluttered
top_ratings = df['rating'].value_counts().head(10)
sns.barplot(x=top_ratings.index, y=top_ratings.values, palette='Reds_r')
plt.title('Top 10 Content Ratings on Netflix', fontsize=14, fontweight='bold')
plt.xlabel('Age Rating', fontsize=12)
plt.ylabel('Number of Titles', fontsize=12)
plt.show()
```

```python


plt.figure(figsize=(10, 6))
# We filter for titles released after 2000 to zoom in on the modern era
recent_content = df[df['release_year'] >= 2000]
sns.histplot(recent_content['release_year'], bins=22, kde=True, color='#E50914')
plt.title('Explosion of Netflix Content (2000 - Present)', fontsize=14, fontweight='bold')
plt.xlabel('Release Year', fontsize=12)
plt.ylabel('Number of Titles Released', fontsize=12)
plt.show()
```

```python
# chose country for see country specific data analyasis

country_name=str(input("Enter country name"))

filtered_df = df[df['country'].str.contains(country_name, case=False)]
filtered_df.head()
```

##Displaying country specific data chose by the user


```python
# displaying fildered dataset analysis

# pie chart of country specific movies vs tv shows
country_views=filtered_df["type"].value_counts()
# country_views
plt.title(f"Netflix Catalog: Movies vs. TV Shows in {country_name}")
plt.pie(country_views,labels=country_views.index,autopct='%1.1f%%', startangle=90, textprops={'fontsize': 12})
plt.show()
```

```python
# popular ratings in given country
country_ratings=filtered_df["rating"].value_counts()
plt.title(f"Top 10 Content Ratings in {country_name}")
plt.xticks(rotation=45)
plt.xlabel('Age Rating', fontsize=12)
plt.ylabel('Number of Titles', fontsize=12)
sns.barplot(x=country_ratings.index, y=country_ratings.values,)
plt.xlabel('Age Rating', fontsize=12)
plt.ylabel('Number of Titles', fontsize=1)
plt.show()
```

```python
# movie release frequency in the country
fd = filtered_df[filtered_df["type"] == "Movie"]

# Calculate the number of movies released each year and sort by year
movie_releases_counts = fd["release_year"].value_counts().sort_index()

# making graph
plt.figure(figsize=(10, 6))
plt.plot(movie_releases_counts.index, movie_releases_counts.values, marker='o', color='#E50914')
plt.title(f'Movie Release Frequency in {country_name}', fontsize=14, fontweight='bold')
plt.xlabel('Release Year', fontsize=12)
plt.ylabel('Number of Movies', fontsize=12)
plt.grid(True)
plt.show()
```
